```python
import numpy as np, pandas as pd, sklearn, seaborn
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
import matplotlib.pyplot as plt

california_housing = fetch_california_housing()
boston_housing = pd.read_csv("BostonHousing.csv")

df_california = pd.DataFrame(california_housing.data, columns=california_housing.feature_names)
df_california['MEDV'] = california_housing.target

df_boston = pd.DataFrame(boston_housing)

normalized_df_boston=(df_boston-df_boston.min())/(df_boston.max()-df_boston.min())

# seaborn.boxplot(data=normalized_df)
# corr = normalized_df.corr()
# seaborn.heatmap(corr)
# seaborn.kdeplot(data=df_boston['medv'])
# seaborn.kdeplot(data=normalized_df['medv'])

X1 = normalized_df_boston[['nox', 'rm', 'dis', 'ptratio', 'lstat']]
Y1 = normalized_df_boston['medv']

x_train, x_test, y_train, y_test = train_test_split(X1, Y1, train_size=0.33, random_state=5)

regressor = LinearRegression()
regressor.fit(x_train, y_train)
y_pred = regressor.predict(x_test)

df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
df

# x1 = np.array(x_test['nox'])
# y1 = np.array(y_pred)
# m,b = np.polyfit(x1,y1,1)

# plt.plot(x_test['nox'], m*x_test['nox'] + b)
# plt.plot(x_test['nox'], y_pred, 'o')

x2 = np.array(x_test['rm'])
y2 = np.array(y_pred)
m,b = np.polyfit(x2,y2,1)

plt.plot(x_test['rm'], m*x_test['rm'] + b)
plt.plot(x_test['rm'], y_pred, 'o')

seaborn.pairplot(normalized_df_boston, x_vars=['nox', 'rm', 'dis', 'ptratio', 'lstat'], y_vars=['medv'], height=9, aspect=0.6, kind='reg')

# print(df_california.keys())
print(df)
print(metrics.mean_absolute_error(y_test, y_pred))
print(metrics.mean_squared_error(y_test, y_pred))
print(np.sqrt(metrics.mean_absolute_error(y_test, y_pred)))
print(metrics.r2_score(y_test, y_pred))
```
